#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
temp2 = temp.groupby('State', as_index=False).agg({'Temperature':np.mean})
temp2.rename(columns = {'Temperature':'Temp_mean'}, inplace=True)

temp3 =  temp.groupby('State', as_index=False).agg({'Temperature':np.std})
temp3.rename(columns = {'Temperature':'Temp_std'},inplace=True)

temp2['Temp_std'] = temp3['Temp_std']
temp2
```
